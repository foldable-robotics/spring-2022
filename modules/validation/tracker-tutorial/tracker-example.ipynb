```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import scipy.interpolate as si
```

```python
df=pd.read_csv(r'pendulum.csv', sep=',')
```

```python
#x = df.x.astype('float64').to_numpy()
x = df.x.to_numpy()
y = df.y.to_numpy()
t = df.t.to_numpy()
```

```python
t_sample = (t[-1]-t[0])/len(t)
t_sample
```

```python
xy = numpy.array([x,y]).T
f = si.interp1d(t,xy.T,fill_value='interpolate',kind='linear')
new_t = numpy.r_[0:t[-1]:t_sample/2]

plt.figure()
a2 = plt.plot(new_t,f(new_t)[0],'bx-')
a1 = plt.plot(t,x,'ro')
plt.legend(a1+a2,['data','interpolation'])
```
