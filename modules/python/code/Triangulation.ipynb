# Delaunay Triangulation

```python
%matplotlib inline
```

```python
from scipy.spatial import Delaunay
import numpy.random
import matplotlib.pyplot as plt
plt.ion()
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
```

```python
a = numpy.random.randn(100,2)
d = Delaunay(a[:,:2])
z = 1-(a**2).sum(1)
```

```python
fig = plt.figure()
for item in a[d.vertices]:
    plt.fill(item[:,0],item[:,1],color = (0,1,0,.1))    

for item in a[d.convex_hull]:
    plt.plot(item[:,0],item[:,1],'r-')

plt.plot(a[:,0],a[:,1],'b.')
plt.axis('equal')
```

```python
fig = plt.figure()    
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(a[:,0],a[:,1],d.vertices,z,cmap=cm.jet,edgecolor = (0,0,0,0))
```
